```python
# condensed columns 
bill_data['item_brand'] = bill_data['SGRP'] + ' ('+ bill_data['BRD'] + ')'

bill_data['MONTH'] = bill_data['MONTH'].map({'M1':'01','M2':'03','M3':'05'})
```

```python
# unique (item + brand)
bill_data.item_brand.nunique()
```

```python
def day_map(x):
    '''Edit the date to same format by adding 0 infront of 
    the number if number less than 10'''
    if x < 10:
        x = '0'+str(x)
        return x
    else:
        x = x
        return x
```

```python
# Convert date number to proper format, then merge date and month , then to_datetime
bill_data['DAY'] = bill_data['DAY'].apply(lambda x : day_map(x))
bill_data['Date'] = bill_data['DAY'].astype(str) +'-' +bill_data['MONTH'] +'-' +'22'
bill_data['Date'] = pd.to_datetime(bill_data['Date'])
```

```python
# Create the index to run for loop
bill_data['trans_index'] = list(zip(bill_data['BILL_ID'],bill_data['Date']))
# keep only unique index
trans_index = bill_data['trans_index'].unique()
# set 2 columns as index             
bill_data.set_index(['BILL_ID','Date'],inplace=True)
```

```python
# Get list of item in each unique index assign to list
shop_list = []
# loop through transaction_index
for index in trans_index:
    #strip each item trailing and head for whitespace out, and assign into shop_list
    shop_list.append([item.strip() for item in bill_data['item_brand'].loc[index].values.tolist()])
```

### Prepare to do a MBA

```python
# Dummy
item_trans = transaction['item'].str.get_dummies('|')

# Change astype to 'bool'
item_trans_bool = item_trans.astype(bool)
```

```python
# frequent_itemsets = apriori(basket, min_support = 0.05, use_colnames = True)

# 5 % of items shown will be selected
frequent_itemsets = fpgrowth(item_trans_bool, min_support = 0.0001, use_colnames = True)

# Show the number of each items (ratio * total rows)
# frequent_itemsets['support_n']=frequent_itemsets['support']*len(basket)

# Sorted
frequent_itemsets.sort_values('support', ascending=False)
```

#### **Association rule** is a method of finding interesting relationships between the variables in a large dataset.

#### Evaluation metrics


```python
# สร้าง association rules
# metrices = ['support', 'confidence', 'lift', 'leverage', and 'conviction']
metric='support'
# rules = association_rules(frequent_itemsets, metric = "support", min_threshold = .2)
# rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = .5)
rules = association_rules(frequent_itemsets, metric = metric, min_threshold = 0.000052)

rules.sort_values(by = 'lift', ascending = False, inplace = True)

# add support number
rules['support_n'] = rules['support']*len(item_trans_bool)

rules.head(2)
```

- antecedent support (LHS support): the ratio of A item
- consequents support (RHS support): the ratio of B item
- support : the ratio of having A and B in the data
- confidence : depends on A --> B or B --> A 
    - if A--> B : when buying A how many will buying B
    
- lift(A-->B) = lift (B-->A) : how many time will buying B if A and B was bought
    - Ex. There is likelihood to buy whole milk and yogurt together 0.82 times (lift) more than buy only whole milk/ only yogurt 


### How to call the name of item from association rules

```python
rules[rules['antecedents'] == 'BLENDED (EVEREST)']
# Call like this will be wrong, !! the values in antecedent support, consequents support are in **set format**
```

```python
# Try the result
item_name = 'KETCHUPS/SAUCES (KISSAN FRESH KETCHUP)'
lift_threshold = 1

rec_result = rules[(rules['antecedents'] == {item_name}) & (rules['lift'] > lift_threshold)]
rec_result
```
